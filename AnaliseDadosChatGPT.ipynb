```python
# Importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o conjunto de dados do CSV
url = "https://raw.githubusercontent.com/labulatif/public/main/adultos.csv"
dados = pd.read_csv(url)

# Visualizar as primeiras linhas do DataFrame
print(dados.head())

# Resumo estatístico
print(dados.describe())

# Visualização de dados: Histogramas das variáveis numéricas
dados.hist(figsize=(12, 10), bins=20)
plt.show()

# Visualização de dados: Gráfico de dispersão
sns.pairplot(dados, hue='income', palette='Set2')
plt.show()

# Tratamento de dados ausentes
print("Valores ausentes:")
print(dados.isnull().sum())

# Análise de distribuição da variável target 'income'
plt.figure(figsize=(8, 6))
sns.countplot(x='income', data=dados, palette='viridis')
plt.title('Distribuição da Variável Target (Income)')
plt.show()
```
